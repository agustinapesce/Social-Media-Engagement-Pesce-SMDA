```python
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

selenium.__version__
```

```python
options = Options()
options.add_experimental_option("detach", True)
driver = webdriver.Chrome(service = Service(ChromeDriverManager().install()),
                            options = options)

# URL of the page containing the embedded content
mediabias_link = 'https://www.allsides.com/media-bias/media-bias-chart'
mediabias_link = "https://infogram.com/1pkr3we7wqklwph9rppgn6vv0vf3nddnkg7"

# Load the page using Selenium
driver.get(mediabias_link)

# Wait for the embedded content to load (you might need to adjust the time depending on the page)
driver.implicitly_wait(10)  # Wait for 10 seconds

# Get the content of the embedded iframe
iframe = driver.find_elements('class', 'InfographicEditor-Contents-Item _24YLdmzyGki1VEeY3thxsm')
iframe_content = iframe.get_attribute('src')

# Load the iframe content using requests
iframe_response = requests.get(iframe_content)
iframe_soup = BeautifulSoup(iframe_response.content, 'html.parser')

driver.quit()
```
